```r
library(tidyverse)
library(ggplot2)
```

```r
twitch_data <- read_csv("twitchdata-update.csv")
colnames(twitch_data) <- make.names(colnames(twitch_data), unique = TRUE)
twitch_data
```

```r
dt = sort(sample(nrow(twitch_data), nrow(twitch_data)*.7))
training_data <- twitch_data[dt,]
testing_data <- twitch_data[-dt,]
```

```r
nrow(training_data)
```

```r
nrow(testing_data)
```

```r
options(repr.width = 42, repr.height = 6)

variable_distribution <- training_data |>
    ggplot(aes(x = Followers)) +
    geom_histogram() +
    labs(x = "Followers", title = "Distrubution Plot") 

    
variable_distribution
```
